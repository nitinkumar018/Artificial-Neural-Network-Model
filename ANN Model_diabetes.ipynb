```python
import numpy as np
```

```python
dataset = np.loadtxt(r"pima-indians-diabetes.data.csv", delimiter= ",")
```

```python
dataset   # matrix
```

```python
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
X
```

```python
Y
```

```python
len(dataset)
```

```python
from keras.models import Sequential
model = Sequential() # Neural Network
```

```python
#from tensorflow.keras.models import Sequential   # another way 
```

```python
#model1 = Sequential()
```

```python
#model1
```

```python
model
```

```python
# structure of model
from keras.layers import Dense
model.add(Dense(20,input_dim = 8, activation = 'relu'))   # 20 input layers (each input layer has 8 neurons)
```

```python
model.add(Dense(10, activation= 'relu'))   # hidden layers
model.add(Dense(10, activation= 'relu'))
```

```python
model.add(Dense(1,activation='sigmoid'))  # output layer
```

```python
model.compile(loss= 'binary_crossentropy', optimizer = 'adam', metrics= ['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X,Y, validation_split = 0.20, epochs= 250, batch_size = 1)
```

```python
history
```

```python
scores = model.evaluate(X,Y)
```

```python
scores
```

```python
scores = model.evaluate(X,Y)
print("%s:%.2f%%"%( model.metrics_names[1],scores[1]*100))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['accuracy']) # accuracy of model on training data
```

```python
plt.plot(history.history['accuracy']) # accuracy of model on training data
plt.plot(history.history['val_accuracy']) # accuracy of model on testing data
plt.title('model accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['train', 'test'], loc = 'upper left')
plt.show()
```

```python
plt.plot(history.history['loss']) # training loss
plt.plot(history.history['val_accuracy']) # testing loss
plt.title('model loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'test'], loc = 'upper left')
plt.show()
```
